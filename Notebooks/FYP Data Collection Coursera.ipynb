```python
# -*- coding: utf-8 -*-
# --- Cell 1: Imports and Setup ---

import time
import re
from datetime import datetime
import pandas as pd
import unicodedata

# Selenium Imports
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By # Import By for locating elements
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, ElementClickInterceptedException, StaleElementReferenceException, WebDriverException

# Webdriver Manager Imports
from webdriver_manager.chrome import ChromeDriverManager

print("[Setup] Importing libraries...")

# --- Configuration ---

# !!! Using ALL Coursera URLs !!!
QUORA_LOG_URLS = [
    'https://www.quora.com/Is-a-Coursera-certificate-useful/log',
    'https://www.quora.com/Are-online-courses-by-Coursera-worth-it/log'
    # ... (and so on) ...
]
MAX_ANSWERS_PER_URL = 100 # Keep limit per URL
# !!! Setting specific output filename for Coursera !!!
OUTPUT_FILENAME = 'coursera_reviews_log.csv'
BASE_URL = 'https://www.quora.com/' # For login
# --- Initialize Data Storage ---
all_answers_data = [] # Initialize list to store data

print("[Setup] Imports and configuration complete.")
print(f"[Setup] Target /log URLs count: {len(QUORA_LOG_URLS)}") # Show count instead of full list
print(f"[Setup] Output file: {OUTPUT_FILENAME}")
print("[Setup] NOTE: This version scrapes /log pages and saves Full Text & Date.")
```

```python
# -*- coding: utf-8 -*-
# --- Cell 2: Function to Initialize Selenium WebDriver ---

def initialize_driver():
    """ Initializes Selenium Chrome WebDriver. """
    try:
        print("[WebDriver] Initializing Chrome WebDriver...")
        options = webdriver.ChromeOptions()
        options.add_argument('--disable-gpu')
        options.add_argument('--no-sandbox')
        options.add_argument('--disable-dev-shm-usage')
        options.add_argument("--disable-notifications") # Try disabling browser notifications
        # Using a realistic user agent might help avoid bot detection
        options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36") # Example: Update to a recent Chrome version
        options.add_experimental_option("excludeSwitches", ["enable-automation"])
        options.add_experimental_option('useAutomationExtension', False)
        service = Service(ChromeDriverManager().install())
        driver = webdriver.Chrome(service=service, options=options)
        driver.maximize_window()
        # Make Selenium less detectable
        driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")
        print("[WebDriver] WebDriver initialized successfully.")
        return driver
    except WebDriverException as e:
         # Catch specific WebDriverException
         print(f"[WebDriver Error] WebDriver setup issue: {e}. Ensure Chrome/Chromedriver is installed and compatible.")
         return None
    except Exception as e:
        print(f"[WebDriver Error] Error initializing WebDriver: {e}")
        return None
```

```python
# -*- coding: utf-8 -*-
# --- Cell 4: Extract Full Text Function (Log Page) - REPAIRED ---

def extract_full_text_log(answer_container_element, driver, wait):
    """
    Extracts the full answer text AND the revision date/time string
    from within a given answer container WebElement on a /log page,
    repeatedly clicking 'more' buttons until none are found.

    Args:
        answer_container_element (WebElement): The Selenium WebElement for the answer container (the wrapper for one revision).
        driver (webdriver): The Selenium WebDriver instance.
        wait (WebDriverWait): The WebDriverWait instance.

    Returns:
        tuple: (full_answer_text, date_str)
               full_answer_text (str): The extracted full answer text, or None if not found/error.
               date_str (str): The extracted date string, or 'N/A' if not found/error.
    """
    full_answer_text = None
    date_str = 'N/A' # Default date
    max_clicks = 5 # Safety limit to prevent infinite loops

    # --- Using PROPOSED GENERAL Selectors (Based on user feedback analysis) ---
    bordered_text_box_selector = "div.q-box.qu-borderRadius--medium.qu-p--medium.qu-mb--medium.qu-borderAll" # Selector for the box containing the text
    more_button_selector = "div.qt_read_more" # Selector for the 'more' button WITHIN the bordered box
    text_paragraph_selector = "p.q-text" # Selector for paragraphs WITHIN the bordered box

    # --- Selector for the Date/Timestamp Area ---
    # This targets the div containing the revision link, buttons, and timestamp
    metadata_footer_selector = "div.q-text.qu-dynamicFontSize--small.qu-mt--small.qu-color--gray_light.qu-passColorToLinks"
    # This targets the specific spans within the footer; the date seems to be the last one.
    footer_spans_selector = "span.c1h7helg" # Class applied to spans containing revision link, buttons, and date

    try:
        # --- Try to Extract Date First (as it's usually outside the bordered box) ---
        try:
            metadata_footer = answer_container_element.find_element(By.CSS_SELECTOR, metadata_footer_selector)
            footer_spans = metadata_footer.find_elements(By.CSS_SELECTOR, footer_spans_selector)
            if footer_spans:
                # Assume the date is the text of the last relevant span
                potential_date = footer_spans[-1].text.strip()
                # Basic check to see if it looks like a date (you might need more robust parsing later)
                if re.search(r'\d{4}', potential_date) and any(month in potential_date for month in ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]):
                     date_str = potential_date
                     # print(f"      [Date] Found: {date_str}") # Optional Debug
                # else:
                     # print(f"      [Date Warn] Last span text '{potential_date}' didn't look like a date.") # Optional Debug
            # else:
                 # print("      [Date Warn] No footer spans found.") # Optional Debug
        except NoSuchElementException:
            print(f"      [Date Warn] Metadata footer not found using selector: '{metadata_footer_selector}'. Date will be N/A.")
        except Exception as date_err:
            print(f"      [Date Error] Error extracting date: {date_err}")


        # --- Now handle Text Extraction (including 'more' button) ---
        # 1. Find the main bordered text box
        try:
            bordered_text_box = answer_container_element.find_element(By.CSS_SELECTOR, bordered_text_box_selector)
        except NoSuchElementException:
            # Check if it's just a simple deleted message (no bordered box)
            try:
                 deleted_text_container = answer_container_element.find_element(By.CSS_SELECTOR, "div.q-text.qu-color--gray")
                 deleted_text = deleted_text_container.text.strip()
                 if "deleted by" in deleted_text.lower():
                      print(f"      [Info] Skipping text extraction for simple deleted message container: '{deleted_text[:50]}...'")
                      return None, date_str # Return None for text, but keep potentially found date
                 else:
                      print(f"      [Text Extract Error] Bordered text box not found and not a recognized deleted message.")
                      return None, date_str # Return None for text, keep date
            except NoSuchElementException:
                  print(f"      [Text Extract Error] Neither bordered text box nor simple deleted message found.")
                  return None, date_str # Return None for text, keep date

        # Click "more" buttons if they exist WITHIN the bordered text box
        clicks = 0
        while clicks < max_clicks:
            try:
                more_button = WebDriverWait(bordered_text_box, 2).until(
                    EC.element_to_be_clickable((By.CSS_SELECTOR, more_button_selector))
                )
                if more_button and more_button.is_displayed():
                    # print(f"      [Expand] Found '(more)' button #{clicks+1}. Clicking...") # DEBUG
                    try:
                        driver.execute_script("arguments[0].scrollIntoView({block: 'center', inline: 'nearest'});", more_button)
                        time.sleep(0.5)
                        more_button.click(); time.sleep(1); clicks += 1
                    except ElementClickInterceptedException:
                        # print("      [Expand Warn] Standard click intercepted. Trying JS click.") # DEBUG
                        try:
                            driver.execute_script("arguments[0].click();", more_button); time.sleep(1); clicks += 1
                        except Exception as js_click_err:
                            # print(f"      [Expand Error] JS click failed: {js_click_err}. Stopping.") # DEBUG
                            break
                    except StaleElementReferenceException: break # Assume done
                    except Exception as click_err:
                        # print(f"      [Expand Error] Clicking '(more)': {click_err}. Stopping.") # DEBUG
                        break
                else: break
            except TimeoutException: break
            except NoSuchElementException: break
            except StaleElementReferenceException: break
            except Exception as loop_err:
                # print(f"      [Expand Error] Loop error: {loop_err}. Stopping.") # DEBUG
                break
        # if clicks == max_clicks: print("      [Expand Warn] Reached max clicks limit.") # DEBUG


        # 3. Extract Full Text from the bordered text box AFTER clicks
        try:
            # Attempt to re-find the box
            try:
                current_bordered_text_box = answer_container_element.find_element(By.CSS_SELECTOR, bordered_text_box_selector)
            except (NoSuchElementException, StaleElementReferenceException):
                current_bordered_text_box = bordered_text_box # Use last known reference

            text_paragraphs = current_bordered_text_box.find_elements(By.CSS_SELECTOR, text_paragraph_selector)
            text_parts = [p.text for p in text_paragraphs if p.text and p.text.strip()]
            if text_parts:
                 full_answer_text = "\n".join(text_parts).strip()
            else:
                 full_answer_text = current_bordered_text_box.text.strip()

            if full_answer_text:
                 full_answer_text = full_answer_text.replace('(more)', '').strip()
                 if not full_answer_text: full_answer_text = None
        except NoSuchElementException:
            print(f"      [Text Extract Error] Could not find paragraphs/text within bordered box.")
            full_answer_text = None
        except StaleElementReferenceException:
             print("      [Text Extract Error] Stale element reference extracting text.")
             full_answer_text = None
        except Exception as text_err:
            print(f"      [Text Extract Error] Extracting text: {text_err}")
            full_answer_text = None

    except StaleElementReferenceException:
        print("      [Error] Stale element reference on main answer container. Skipping.")
        return None, date_str # Return None for text, keep potentially found date
    except Exception as e:
        print(f"      [Error] General error in extract_full_text_log: {e}")
        return None, date_str # Return None for text, keep potentially found date

    # Return both text and date
    return full_answer_text, date_str
```

```python
# -*- coding: utf-8 -*-
# --- Cell 5: Main Scraping Logic (Log Page Hybrid) ---

# Ensure the 'driver' variable from Cell 3 exists and is usable
if 'driver' in locals() and driver:
    # Define wait object here for use in the loop and function
    wait = WebDriverWait(driver, 15) # Increased main wait slightly

    try:
        print("\n--- Starting Data Extraction Loop (Log Page Hybrid: Full Text Only) ---")
        # Loop through target /log URLs from Cell 1
        for url in QUORA_LOG_URLS: # Use the /log URLs
            print(f"\n[INFO] Navigating to: {url}")
            processed_count_for_url = 0 # Track processed answers for this specific URL
            try:
                driver.get(url)
                # Wait for the main content area using the reliable ID selector
                log_page_content_selector = "#mainContent" # <-- Using reliable ID selector
                try:
                    # Wait for the element to be present AND visible
                    wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, log_page_content_selector)))
                    print(f"  [INFO] Log page content area ('{log_page_content_selector}') loaded.")
                except TimeoutException:
                    print(f"  [WARN] Timed out waiting for main log page content ('{log_page_content_selector}'). Page might be empty, structure changed, or login failed. Skipping URL.")
                    continue # Skip to next URL if page doesn't load properly

                # --- Scrolling (Using "Unlimited" scroll based on height change) ---
                print("  [INFO] Waiting and scrolling /log page (will stop when height stops changing)...")
                scroll_pause_time = 3
                
                last_height = driver.execute_script("return document.body.scrollHeight")
                no_change_count = 0
                scroll_attempt = 0

                while True: # Loop indefinitely until break condition
                    scroll_attempt += 1
                    # print(f"    Scrolling down (Attempt #{scroll_attempt})...") # Optional Debug
                    try:
                        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                        time.sleep(scroll_pause_time)
                        new_height = driver.execute_script("return document.body.scrollHeight")
                    except Exception as scroll_exec_err:
                         print(f"      [Scroll Error] Could not execute scroll script: {scroll_exec_err}")
                         time.sleep(scroll_pause_time)
                         continue

                    if abs(new_height - last_height) < 10:
                        no_change_count += 1
                        # print(f"    Scroll height changed little ({no_change_count} time(s))...") # Optional
                    else:
                        no_change_count = 0
                    last_height = new_height

                    if no_change_count >= 3:
                         print(f"    Scrolling stopped after {scroll_attempt} attempts.")
                         break
                        
                    if scroll_attempt > 100: # Safety break
                        print(f"    [WARN] Scrolling stopped after {scroll_attempt} attempts (safety break).")
                        break

                print("  [INFO] Finished scrolling.")
                time.sleep(3) # Final wait after scrolling

                # --- Find Answer Containers ---
                # Using the :has() selector - Requires modern Chrome/ChromeDriver
                answer_container_selector = "#mainContent > div > div > div > div.q-box:has(> div.q-box.qu-borderRadius--medium.qu-p--medium.qu-mb--medium.qu-borderAll)" # <-- :has() SELECTOR

                print(f"  [INFO] Finding answer containers using selector: '{answer_container_selector}'")
                answer_container_elements = [] # Initialize
                try:
                    time.sleep(1)
                    answer_container_elements = driver.find_elements(By.CSS_SELECTOR, answer_container_selector)
                    print(f"  [INFO] Found {len(answer_container_elements)} potential answer container elements (using :has()).")
                except NoSuchElementException:
                     print(f"  [WARN] No answer containers found with selector: '{answer_container_selector}'")
                except Exception as find_err:
                    print(f"  [ERROR] Could not find answer containers: {find_err}")
                    print("  [Hint] Ensure your Chrome browser & ChromeDriver support CSS :has() or check selector syntax.")
                    answer_container_elements = []

                if not answer_container_elements:
                     print(f"  [WARN] No answer elements found matching the :has() criteria for {url}. Skipping.")
                     continue

                # --- Loop through found elements ---
                for i, answer_element in enumerate(answer_container_elements):
                    if processed_count_for_url >= MAX_ANSWERS_PER_URL:
                        print(f"    [INFO] Reached max answers limit ({MAX_ANSWERS_PER_URL}) for this URL.")
                        break

                    print(f"  [INFO] Processing Answer Element #{i+1} (Matched Container)")
                    try:
                         driver.execute_script("arguments[0].scrollIntoView({block: 'center', inline: 'nearest'});", answer_element)
                         time.sleep(0.3)
                    except Exception as scroll_err:
                         print(f"      [Warn] Could not scroll element {i+1} into view: {scroll_err}")

                    # --- Extract Data ---
                    # unpacks TWO values: text and date
                    full_text, extracted_date = extract_full_text_log(answer_element, driver, wait)

                    # --- Store Data ---
                    # Only store if full_text is not None (function might return None for text even if date is found, e.g., for deleted messages)
                    if full_text:
                        # Keyword check for deleted answers remains useful
                        if "quora deleted this answer" in full_text.lower() or \
                           "author deleted this answer" in full_text.lower() or \
                           "deleted by quora moderation" in full_text.lower():
                            print(f"    [INFO] Skipping element #{i+1}: Detected as likely deleted answer text.")
                        else:
                            # !!! Store the extracted_date instead of 'N/A' !!!
                            print(f"    [OK] Extracted: Date='{extracted_date}' Text='{str(full_text)[:80].replace(chr(10),' ')}...'") # Show date in log
                            all_answers_data.append({
                                'Question_Log_URL': url,
                                'Full_Answer_Text': full_text,
                                'Date': extracted_date # Use the extracted date here
                            })
                            processed_count_for_url += 1
                    # else: # Element was skipped (e.g., simple deleted msg) or text extraction failed within function
                         # print(f"    [INFO] Skipping element #{i+1} as no valid text was returned.")


                if processed_count_for_url == 0:
                    print(f"  [WARN] Found {len(answer_container_elements)} containers matching :has(), but extracted no valid text from {url}. Check selectors inside Cell 4.")

            except WebDriverException as wde:
                 print(f"  [ERROR] WebDriver error processing page {url}: {wde}. Attempting to continue...")
            except Exception as page_error:
                print(f"  [ERROR] General error processing page {url}: {page_error}")

            # Polite delay
            page_wait_time = 5
            print(f"  [INFO] Waiting {page_wait_time} seconds before next URL...")
            time.sleep(page_wait_time)

    except Exception as main_error:
        print(f"[ERROR] An unexpected error occurred in the main scraping loop: {main_error}")

    finally:
        # --- Close the browser ---
        if 'driver' in locals() and driver:
            print("\n[WebDriver] Closing WebDriver...")
            try:
                driver.quit()
                print("[WebDriver] WebDriver closed.")
            except Exception as quit_err:
                print(f"[WebDriver Error] Error while quitting driver: {quit_err}")
            finally:
                 driver = None
        else:
             print("[WebDriver] WebDriver was not active or already closed.")

else:
    print("\n[Error] Cannot proceed: WebDriver not initialized or manual login step failed/skipped.")

# --- Final Summary ---
# (Same as before)
print(f"\n--- Scraping Finished ---")
print(f"Total answers collected: {len(all_answers_data)}")
if all_answers_data:
    print("\nPreview of collected data:")
    try:
        # Ensure Date column is included if data exists
        cols = ['Question_Log_URL', 'Full_Answer_Text', 'Date']
        df_preview = pd.DataFrame(all_answers_data)
        # Reorder/select columns, handling potential missing 'Date' key if all extractions failed weirdly
        df_preview = df_preview[[c for c in cols if c in df_preview.columns]]
        print(df_preview.head())
    except Exception as df_err:
        print(f"[Error] Could not create DataFrame preview: {df_err}")
else:
    print("\nNo data was collected.")
```

```python
# -*- coding: utf-8 -*-
# --- Cell 5.5: Filter Data by Year ---

print("\n--- Filtering Data by Year (2024-2025) ---")

filtered_answers_data = []
parse_errors = 0
kept_count = 0
skipped_count = 0

# Ensure all_answers_data exists and is a list
if 'all_answers_data' in locals() and isinstance(all_answers_data, list):
    print(f"Processing {len(all_answers_data)} records...")
    for item in all_answers_data:
        date_str = item.get('Date', 'N/A') # Get the date string, default to 'N/A'
        year = None

        if date_str != 'N/A':
            try:
                # Attempt to parse the date string
                # Format: "April 16, 2025 at 2:00:43 AM" -> "%B %d, %Y at %I:%M:%S %p"
                datetime_obj = datetime.strptime(date_str, "%B %d, %Y at %I:%M:%S %p")
                year = datetime_obj.year
            except ValueError:
                # Handle cases where the date string doesn't match the expected format
                print(f"  [Warn] Could not parse date string: '{date_str}'")
                parse_errors += 1
            except Exception as e:
                print(f"  [Error] Unexpected error parsing date '{date_str}': {e}")
                parse_errors += 1

        # Check if the year is 2024 or 2025
        if year is not None and year in [2024, 2025]:
            filtered_answers_data.append(item)
            kept_count += 1
        else:
            skipped_count += 1
            # Optional: Print skipped dates for debugging
            # if year is not None:
            #     print(f"  [Info] Skipping item with date: {date_str} (Year: {year})")
            # elif date_str != 'N/A':
            #     pass # Already warned about parse error
            # else:
            #     print(f"  [Info] Skipping item with date: N/A")


    print(f"\nFiltering Complete:")
    print(f"  - Kept {kept_count} records (Year 2024 or 2025).")
    print(f"  - Skipped {skipped_count} records (Other years or N/A/unparseable date).")
    if parse_errors > 0:
        print(f"  - Encountered {parse_errors} date parsing errors.")
else:
    print("[Error] 'all_answers_data' not found or is not a list. Cannot filter.")
    filtered_answers_data = [] # Ensure it's an empty list if source data is bad


# Now 'filtered_answers_data' holds only the records from 2024 and 2025
# The next cell (Cell 6) should use this list.
```
